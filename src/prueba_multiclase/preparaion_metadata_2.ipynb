```python
#cargamos el dataset
df = pd.read_csv ('/Users/luiseduardogarciablanco/Desktop/nueva data cancer/prueba_data_18_19_20/metadata_18_19_20_completo.csv', index_col=0)
df
```

```python
import pandas as pd

# Suponiendo que tu dataset está cargado en un DataFrame llamado 'df'
print("Número de columnas:", df.shape[1])
print("Nombres de las columnas:", df.columns)

# Contar valores NaN por columna
print("Valores NaN por columna:")
print(df.isnull().sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Suponiendo que tu DataFrame se llama 'df'
# Obtener las columnas con valores NaN
cols_with_nan = df.columns[df.isnull().any()]

# Iterar sobre cada columna con NaN
for col in cols_with_nan:
    # Crear una figura
    plt.figure(figsize=(10, 5))

    # Contar los valores no nulos y nulos
    values_counts = df[col].isnull().value_counts()

    # Crear el gráfico de barras
    plt.bar(['No NaN', 'NaN'], values_counts, color=['blue', 'red'])

    # Personalizar el gráfico
    plt.xlabel('Valor')
    plt.ylabel('Cantidad')
    plt.title(f'Distribución de valores en la columna {col}')

    # Mostrar el gráfico
    plt.show()
```

```python
# Eliminar filas con cualquier valor NaN en el DataFrame
df_sin_NaN = df.dropna()
df_sin_NaN
```

```python
# Convertir la columna age_approx a enteros
df_sin_NaN['age_approx'] = df_sin_NaN['age_approx'].astype(int)
df_sin_NaN
```

```python
# Aplicar One Hot Encoding a las columnas 'sex' y 'anatom_site_general'
df_encoded = pd.get_dummies(df_sin_NaN, columns=['sex', 'anatom_site_general'])
df_encoded
```

```python
# Convertir True a 1 y False a 0 en todo el DataFrame
df_encoded = df_encoded.astype(int)
df_encoded
```

```python
#guardamos el dataset

import pandas as pd

# Directorio y nombre del archivo
directorio = '/Users/luiseduardogarciablanco/Desktop/nueva data cancer/prueba_data_18_19_20'
nombre_archivo = 'metadata_18_19_20_sin_NaN_code.csv'

# Ruta completa del archivo
ruta_archivo = f'{directorio}/{nombre_archivo}'

# Guardar el DataFrame en un archivo CSV manteniendo el índice
df_encoded.to_csv(ruta_archivo, index=True)
```
